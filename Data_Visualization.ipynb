```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
pd.read_csv('V3_Data_1997.csv')
df = pd.read_csv('V3_Data_1997.csv')

date = np.array(df.DATE.values)
house = np.array(df.Housing_Price_Index.values)
date = np.linspace(1997,2021,97)
plt.plot(date, house, label = 'Housing Price Index')
plt.legend()
plt.xlabel('Date')
plt.show()

plt.plot(date, df.Unemp_Rate.values, label = 'Unemployment Rate')
plt.legend()
plt.xlabel('Date')
plt.show()

plt.plot(date, df.Med_Income_SL.values, label = 'Median Income in Salt Lake County')
plt.legend()
plt.xlabel('Date')
plt.show()

plt.plot(date, df.Population.values, label = 'Population')
plt.legend()
plt.xlabel('Date')
plt.show()

plt.plot(date, df.Rental_Vac_Rate.values, label = 'Rental Vaccancy Rate')
plt.legend()
plt.xlabel('Date')
plt.show()
```
