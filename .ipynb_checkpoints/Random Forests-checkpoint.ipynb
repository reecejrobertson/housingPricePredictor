```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

### No forward differences

```python
rfc = RandomForestRegressor()
df = pd.read_csv('Output.csv').set_index('DATE')
features = list(df.columns)
features.remove('UTSTHPI')
idx = df.index
X = df[features]
y = df['UTSTHPI']
```

```python
rfc.fit(X, y)
list(zip(X.columns, rfc.feature_importances_))
```

### First forward difference

```python
df = pd.read_csv('Output.csv').set_index('DATE')
cols = df.columns
idx = df.index
rfc = RandomForestRegressor()
data = np.array(df)[1:]- np.roll(np.array(df), 1, axis=0)[1:]
df = pd.DataFrame(data)
df.columns = cols
df.index = idx[1:]
X = df[features]
y = df['UTSTHPI']
```

```python
rfc.fit(X, y)
list(zip(X.columns, rfc.feature_importances_))
```

### No forward difference with normalization

```python
rfc = RandomForestRegressor()
df = pd.read_csv('Output.csv').set_index('DATE')
df = (df - df.mean()) / df.std()
features = list(df.columns)
features.remove('UTSTHPI')
idx = df.index
X = df[features]
y = df['UTSTHPI']
```

```python
rfc.fit(X, y)
list(zip(X.columns, rfc.feature_importances_))
```
